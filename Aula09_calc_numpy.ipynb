```python
import numpy as np # importando a biblioteca numpy
```

```python
notas = [5, 6, 7, 8, 9] # criando uma lista de notas
```

```python
media = np.mean(notas) # calculando a média da lista de notas
# mean é média.
print(f'A média é {media}')
```

```python
desvio = np.std(notas, ddof=1) # calculando o desvio padraão da lista de notas
# std é standard desviation que é o desvio padrão
# ddof=1 desvio padrão da amostra
print(desvio)
```

Exercício:
Elabore um programa que receba do usuário uma string com n notas e a transforme em uma lista com as notas separadas e tipadas como float, e após isso calcule a média e o desvio padrão, usando a biblioteca Numpy, e no final apresente os valores da média e desvio padrão.

```python
dados = input('Informe a notas de 0 a 10 separadas por espaços: ')
print(dados)
print(type(dados))

notas = dados.split() # cortar/separar os números com vírgula
print(notas)
print(type(notas))
```

```python
for i in range(len(notas)):
  notas[i] = float(notas[i]) # transformar lista em float
```

```python
import numpy as np
media = np.mean(notas)
print(f'A média das notas informadas é: {media:.2f}')
```

```python
desvio = np.std(notas, ddof=1) # degrees of freedom
print(f'O desvio padrão das notas informadas é: {desvio:.2f}')
```
