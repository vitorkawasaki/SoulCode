# Questão 1
**Q1: Declare duas variáveis e realize as quatro operações básicas (soma, subtração, divisão e multiplicação). Ao final, exiba todos os resultados em um print**

```python
print('Vamos calcular!')
n1 = int(input('Digite um número: '))
n2 = int(input('Digite outro número: '))

print('\n')

somar = n1 + n2
diminuir = n1 - n2
dividir = n1 / n2
multi = n1 * n2

print(f'A soma é {somar}, a subtração é: {diminuir}, a divisão é: {dividir}, a multiplicação é: {multi}')
```

# Questão 2

**Q2: Crie uma variável do tipo float chamada comprimento e outra do tipo float chamada largura. Peça para o usuário inserir esses dados, e realize dois cálculos: area e perímetro, ao final, use print para mostrar os resultados.** área = comprimento * largura e perímetro é igual a soma dos lados.

```python
comprimento = float(input('Digite o comprimento: '))
largura = float(input('Digite largura: '))

area = comprimento * largura
perimetro = 2 * (comprimento + largura)

print(f'A área é: {area} m².') # f usar quando tem informação antes e depois do resultado
print('O Perimétro é: ', perimetro) # usar quando

'''
- Sempre separar o texto por partes pequenas para facilitar a programação e tornar isso
um hábito para evitar erros de programação.
Separar passo a passo.
'''
```

# Questão 3
**Escreva um programa em Python que solicite ao usuário o raio da base e a altura de um cilindro. Calcule e imprima o volume do cilindro usando a fórmula: Volume = π * raio^2 * altura. PI = 3,14**

```python
raio_base = float(input('digite o raio da base: '))
altura = float(input('digite a altura: '))
PI = 3.14 #numero constante usar a variável toda em maiúsculo

volume = PI * (raio_base**2) * altura

print('volume: ', volume)
```
