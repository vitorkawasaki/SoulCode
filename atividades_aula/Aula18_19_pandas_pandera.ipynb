##**1 - Estatística Descritiva**

As estatísticas descritivas envolvem os métodos usados para resumir, simplificar e descrever os principais recursos de um conjunto de dados.

Este ramo da estatística visa fornecer uma visão geral concisa e significativa dos dados por meio de medidas de tendência central (média, mediana, moda), dispersão / medidas de variabilidade (intervalos, variância, desvio-padrão) e representação gráfica, tais como histogramas, gráfico de barras, gráfico de setores, gráficos de dispersão e etc.

As estatísticas descritivas nos ajudam a compreender as características fundamentais e de um conjunto de dados e torná-lo mais gerenciáveis para análise.

## **10 - Análise de Dados**

###**10.1 - Gráficos importantes para um Analista de Dados**
* BARRAS (horizontal e vertical - CATEGÓRICAS)
  * Mostra a relação entre uma variável categórica e uma variável numérica.
  * Útil para comparar valores entre diferentes categorias.
    * Exemplos incluem gráficos de barras verticais e horizontais.

* SETORES (pizza/rosquinha - CATEGÓRICAS) 2 a 3 categorias, no máximo 5.
  * Mostra a distribuição de uma variável categórica como uma proporção do todo.
  * Útil para destacar partes de um todo.

* LINHAS (séries temporais)
  * Mostra a mudança de uma variável ao longo do tempo ou em uma sequência ordenada.
  * Útil para destacar tendências e padrões ao longo do tempo.

* DISPERSÃO (numérica: INT ou FLOAT (melhor FLOAT) - o quanto as variáveis são co-relacionadas entre si.
 * Mostra a relação entre duas variáveis numéricas.
 * Útil para identificar correlações e padrões de dispersão nos dados.

* HISTOGRAMA (numéricas contínuas)
  * Representa a distribuição de uma variável numérica em intervalos (bins).
  * Útil para visualizar a distribuição e a forma dos dados.

* BOXPLIT (numéricas discretas) - medidas, localização de concentração de dados
  * Sumariza a distribuição de uma variável numérica e destaca outliers.
  * Mostra a mediana, quartis, intervalo interquartil e outliers potenciais.
  * Útil para identificar a dispersão dos dados e detectar discrepâncias.

* MAPAS (longitude, latitude)
  * Representa dados geográficos em um mapa.
  * Útil para visualizar padrões espaciais e geográficos.

* CORRELAÇÃO (numéricas)
  * Representa a relação entre duas variáveis numéricas.
  * Útil para identificar o grau e a direção da relação entre variáveis.

* GRÁFICOS INTERATIVOS
  * Gráficos que permitem ao usuário interagir com os dados, como zoom, filtro e destaque.
  * Permitem explorar os dados de forma mais dinâmica e personalizada.

* NUVEM DE PALAVRAS
  * Usada para visualizar a frequência de palavras em um texto.
  * As palavras são codificadas com tamanhos diferentes com base na sua frequência.
  * Útil para identificar os principais temas ou tópicos em um texto.
