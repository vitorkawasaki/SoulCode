##**1 - Comando de Controle de Fluxo**
IF, ELSE,ELIF
*  servem para verificar condições.




*  Execução de código consome memória.
*  Comando de controle de fluxo serve para executar apenas uma parte dos códigos para consumir menos memória.

*  Super importante ter o dois pontos ( : ) para funcionar a sintaxe.
*  Identação é o espaço depois dos dois pontos da linha anterior para dizer que a linha de baixo faz parte da linha de cima.(filho)
*  Identação é um espaço colocado pelo python para separar os conteúdos de maneira hierarquica.
*  No python a identação é OBRIGATÓRIA!

```python
idade = 18
if idade >= 18: # condição => True, False
   print("Você é maior de idade")
```

* Estrutura do if

  1.   variavel = int(input('frase'))
  2.   if variavel >= numero:
  3.   print('frase')

* pular linha para a próxima variável

* duas situações usa if e else, acima de duas, usa o elif

  1.  elif é o mesmo que else if
  2.  if faz a primeira vericação.
  3.  elif faz as outras verificações, se houverem.
  4.  else faz a última verificação.

*  duas situações se usa: if e else, acima de duas linhas, usa o elif.
*  o else serve para fechar (apenas o último)

```python
saldo = float(input('Digite o valor do seus saldo: ')) # saldo recebe float
if saldo > 0: # se saldo é maior que zero
  print('O seu saldo é positivo')

if saldo == 0: # se saldo é igual a zero
  print('o seu saldo é neutro')

else: # se não
  print('O seu saldo é negativo')
```

```python
nota = float(input('Digite sua nota: '))

if nota >= 90: # faz uma primeira verificação
  print('A')
elif 80 <= nota < 90:  # faz uma segunda verificação
  print('B')
elif 70 <= nota < 80: # faz uma terceira verificação
  print('C')
elif 60 <= nota < 70: # faz uma quarta verificação
  print('D')
else: # faz a última verificação
  print('F')
```

### Crie um programa Python que tenha if aninhado
*  if dentro de if

```python
# if aninhado
nota = float(input("Digite sua nota: "))
frequencia = int(input("Digite sua frequência: "))

if frequencia >= 75:
  if nota >= 60: # faz parte da mesma família
    print("Aprovado!")
  if nota <= 10: # adicionei informação
    print("Reprovado! Vou te encaminhar para a aula de reforço")
  else:
    print("Reprovado por nota!")
else:
  print("Reprovado por frequência")
```

```python
# identação: uma tecla TAB ou dois espaços
saldo = float(input("Adicione o seu saldo: "))
limite = 1000
# antes de adicionar mais if abaixo do if, melhor fazer if e else no começo e depois
# adicionar entre os dois, para evitar esquecimento
if saldo > 0:
  if saldo >= limite:
    print("Saldo suficiente para compra")
  else:
    print("Saldo insuficiente")
else:
  print("Conta sem saldo!")
```

###Q2: Crie um programa em Python que solicite ao usuário inserir um número inteiro. O programa deve verificar se o número é par e imprimir uma mensagem indicando se é par ou ímpar.

```python
n1 = int(input("digite um número inteiro: ")) # digite um número inteiro
if n1 % 2 == 0 : # verificar se o número é par
  print(f'O seu número {n1} é par.') # mistura de texto e números
  print('o seu número é par') # apenas texto informando o resultado
else:
  print(f'O seu número {n1} é ímpar.') # imprimir => mistura de texto e números
  print("O seu número é ímpar") # apenas texto informando o resultado
```

###Crie um programa que verifica se um número é múltiplo de 3. Se for, avalie se é também divisível por 6.

```python
multi = int(input('digite um número: '))

if multi % 3 == 0:
  print('é múltiplo de 3')
  if multi % 6 == 0:
    print('é multiplo de 6')
  else:
    print('não é divisível por 6')
else:
  print('não')
```

###Q6: Crie um programa em python para verificar se um triângulo é equilátero, isósceles e escaleno. Peça para o usuário digitar o valor dos três lados.**
*  Equilátero tem os 3 lados iguais, Isósceles 2 lados iguais, Escaleno 3 lados diferentes

```python
l1 = float(input('Digite um número: '))
l2 = float(input('Digite outro número: '))
l3 = float(input('Digite outro número: '))

if l1 == l2 == l3:
  print('equilatero')
elif l1 == l2 or l2 == l3 or l1 == l3:
  print('isósceles')
else:
  print('escaleno')
```
