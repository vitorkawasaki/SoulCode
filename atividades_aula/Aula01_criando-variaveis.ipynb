##2 - Criando variáveis

```python
nome = 'Vitor' # variável str - criando a variavel nome e atribuindo a ela o texto Vitor
idade = 37 # variável int - criando a variável idade e atribuindo a ela o valor 37
altura = 1.68 # variável float - Altura recebe 1.68
# o caracter # (hashtag) significa: operador de atribuição

print(nome) # mostrando o conteúdo armazenado na variável nome
print(idade) # mostrando o conteúdo armazenado na variável idade
print(altura)  # mostrando o conteúdo armazenado na variável altura
print(nome, idade, altura) # mostrando na mesma linha o conteúdo de nome, idade e altura
print(f'O aluno {nome} tem {idade} anos de idade e {altura} m de altura') # mostrando o conteúdo de nome e idade combinados com texto livre
```
