## 2 - Gráfico Exploratório
* nível para verificação do gráfico (você mesmo)
* não serve para apresentar para empresas
* evolução temporal de alguma métrica
* um bom gráfico é aquele que conta uma história sme precisar de uma explicação, bate o olho e já sabe o que significa
* precisa de: título, legenda

###2.1 - Gráfico de linhas no Pandas (plot)


```python
df['vendas'].plot()
# df na coluna 'vendas' na função PLOT (linhas)
# gráfico de linha é ótimo para evolução temporal de alguma métrica
```

###2.2 - Gráfico de Histograma (hist)

```python
df['receita'].hist()
# df na coluna 'receita' na função HIST (histograma)
```

###2.3 - Diagrama de Dispersão (scatter)
* ótimo gráfico para enxergar a relação de duas coisas
df.plot.scatter('x', 'y')

```python
df.plot.scatter('vendas', 'lucro')
```

###2.4 - Gráfico de Separatrizes (boxplot)
* Medidas Separatrizes

```python
df['receita'].plot(kind='box')
# em traço horizontal extremo de baixo (LI)
# em traço horizontal extremo de cima (LS)
# a caixa central, na linha de baixo é o 1partil
# a caixa central, na linha do meio é a mediana
# a caixa central, na linha de cima é o 2partil
# 25% entra cada linha horizontal
```

###2.5 - Gráfico de Barras (kind='bar')

```python
df['vendas'].plot(kind='bar')
```

###2.6 - Gráfico de área (linhas)
* evolução de vendas

```python
df['vendas'].plot(kind='area')
# parte de baixo da linha preenchida
```

###2.7 - Gráfico de pizza

```python
df['vendas'].plot(kind='pie')
```
