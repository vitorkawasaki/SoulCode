*  Multiplicador = [7,7,7,7]

*  isso abaixo é matriz.
  *  7 x 1

*  2 dimensões, usar 2 "for". "for' dentro de "for"

```python
for multiplicador in range(7,8):
# somente para imprimir o 7 coloca (7,8), pois não imprime o último
  for multiplicando in range(1,11): # imprimir do 1 ao 10.
    resultado = multiplicador * multiplicando
    print(f'{multiplicador} x {multiplicando} = {resultado}')
```

```python
for multiplicador in range(7,8):
# somente para imprimir o 7 coloca (7,8), pois não imprime o último
  for multiplicando in range(1,11): # imprimir do 1 ao 10.
    for mult in range(2,3): # multiplicar por 2 apenas
      resultado = multiplicador * multiplicando * mult
      print(f'{multiplicador} x {multiplicando} x {mult} = {resultado}')
```

#**2 - While**
*  Necessita do contador
*  Precisa ser incrementado ou decrementado

```python
#normalmente usa palavras maiores que o "for"
contagem = 5 #contador

while contagem > 0: #condição
  print(contagem)
  contagem -= 1 # toda vez que for contar, vai reduzir 1
```

*  muito utilizado em escopo de programas
*  repetição de bloco que já existem

```python
# imprimir apenas números pares
numero_atual = 2
while numero_atual <= 10:
  print(numero_atual)
  numero_atual += 2
```

```python
limite = 10
soma = 0
numero_atual = 1

while numero_atual <= limite: #condição
# ação do código
  soma += numero_atual
  # 0 = 0 + 1
  # 1 = 1 + 2
  # 3 = 3 + 3
  numero_atual += 1
  print(numero_atual)
print(soma)
```

Peça ao usuário para inserir as notas de um aluno. Continue solicitando notas até que o usuário insira um valor negativo. Calcule a média das notas inseridas e imprima o resultado.

```python
soma_notas = 0
qtde_notas = 0
nota = 0

while nota >= 0:
  nota = float(input("Digite a nota do aluno: (Insira um valor negativo para encerrar.) "))
  if nota >= 0:
    soma_notas += nota
    qtde_notas += 1

if qtde_notas > 0: # usado o if aqui, para não pegar a nota = 0.
  media = soma_notas / qtde_notas
  print(f'A média das notas é: {media}')
else:
  print("Nenhuma nota foi inserida")
```

```python
notas = []
nota = 0

while nota >= 0:
  nota = float(input("Digite a nota do aluno: (Insira um valor negativo para encerrar.) "))
  if nota >= 0:
    notas.append(nota)

if len(notas) > 0:
  media = sum(notas) / len(notas) # sum - soma a lista, len - conta a qtde da lista
  print(f'A media das notas é: {media:.2f}')
else:
  print("nenhuma nota foi inserida")
```

##2.3 - Gerador de número aleatório na biblioteca

```python
import random

numero_secreto = random.randint(1,20) # intervalo do 1 ao 20
tentativas = 0
limite_tentativas = 5

while tentativas < limite_tentativas:
  palpite = int(input("Adivinhe o número: "))
  tentativas += 1

  if palpite == numero_secreto:
    print(f'Parabéns! Você acertou em {tentativas} tentativas.')
    break # quebra o fluxo do programa se acertar
  elif palpite < numero_secreto:
    print("Tente um número maior.")
  else:
    print("Tente um número menor")

if tentativas == limite_tentativas:
  print(f"Limite de tentativas atingido. O número era {numero_secreto}")
```

```python
# teste do Rafael

# Inicializando saldo inicial
saldo = 10000.00

# Mensagem de boas-vindas
print("Bem-vindo ao Banco Virtual!")
# Exibindo opções para o usuário
print("\nOpções:")
print("1. Verificar saldo")
print("2. Fazer saque")
print("3. Fazer depósito")
print("4. Sair")

while True:
    # Solicitando a escolha do usuário
    escolha = input("\nEscolha uma opção (1, 2, 3 ou 4): ")

    # Verificando a escolha do usuário
    if escolha == "1":
        print(f"Seu saldo atual é: R${saldo:.2f}")
    elif escolha == "2":
        # Fazendo saque
        valor_saque = float(input("Digite o valor que deseja sacar: "))
        if valor_saque > saldo:
            print("Saldo insuficiente. Operação não realizada.")
        else:
            saldo -= valor_saque
            print(f"Saque de R${valor_saque:.2f} realizado. Saldo atual: R${saldo:.2f}")
    elif escolha == "3":
        # Fazendo depósito
        valor_deposito = float(input("Digite o valor que deseja depositar: "))
        saldo += valor_deposito
        print(f"Depósito de R${valor_deposito:.2f} realizado. Saldo atual: R${saldo:.2f}")
    elif escolha == "4":
        print("Obrigado por utilizar nosso serviço. Volte sempre!")
        break  # Sai do loop quando o usuário escolhe sair
    else:
        print("Opção inválida. Tente novamente.")
```
