# Esquema de Validação

```python
schema = pa.DataFrameSchema({'renda': pa.Column(pa.Float),
                             'criancas': pa.Column(pa.Int),
                             'adolescentes': pa.Column(pa.Int),
                             'retorno': pa.Column(pa.Int),
                             'vinhos': pa.Column(pa.Int),
                             'frutas': pa.Column(pa.Int),
                             'carne': pa.Column(pa.Int),
                             'peixe': pa.Column(pa.Int),
                             'doces': pa.Column(pa.Int),
                             'compras_negocios': pa.Column(pa.Int),
                             'compras_web': pa.Column(pa.Int),
                             'catalogo': pa.Column(pa.Int),
                             'compras_loja': pa.Column(pa.Int),
                             'visita_web': pa.Column(pa.Int),
                             'Response': pa.Column(pa.Int),
                             'idade': pa.Column(pa.Int),
                             'cliente_por_dia': pa.Column(pa.Int),
                             'divorciados': pa.Column(pa.Int),
                             'casados': pa.Column(pa.Int),
                             'solteiros': pa.Column(pa.Int),
                             'casados_juntos': pa.Column(pa.Int),
                             'viuvos': pa.Column(pa.Int),
                             'ensino_medio': pa.Column(pa.Int),
                             'ensino_fundamental': pa.Column(pa.Int),
                             'graduacao': pa.Column(pa.Int),
                             'mestrado': pa.Column(pa.Int),
                             'doutorado': pa.Column(pa.Int),
                             'total': pa.Column(pa.Int),
                             'prod_regular': pa.Column(pa.Int),

                              })
```
