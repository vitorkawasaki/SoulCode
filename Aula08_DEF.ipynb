*  Crie um programa que receba do usuário n valores de notas (0 - 10) calcule a média e informe o usuário a média aritmética dos valores informados Método de resolução com while

```python
notas = []
c = 0 # contador
n = int(input('Informe a quantidade de notas a inserir no sistema: '))

while c < n: # enquanto c for menor n, o sistema vai perguntar ao usuário.
  notas.append(float(input(f'Informe a {c+1}ª nota de 0 a 10: ')))
# armazena com o append. type: float.
# falta a atualização a cada pedido.
  c += 1 # operador incremental. usar depois do append
media = sum(notas)/len(notas) # somar notas. len, contar notas.
print(f'A média das {n} notas é {media:.1f}')
print(f'A média das {len(notas)} notas é {media:.1f}') # consome mais recursos, mas dá certo.
```

* Usando "for" com "range"

```python
dados = input("Informe a nota de 0 a 10 separada por espaços: ") # deixa com string
# com float ele pode juntar os números: 1 2 3 4 5 vai virar 12345.

notas = dados.split() # notas recebe dados.split, deixar vazio, vai deixar o espaço
# transforma os dados em lista.
for i in range(len(notas)): # para i em range (intervalo). contar as notas usa o len(notas)
  notas[i] = float(notas[i]) # transformar lista em float
  # aqui ele grava a informação
```

*  Crie um programa que receba do usuário n valores de notas (0 - 10)
calcule a média e informe o usuário a média aritmética dos valores informados
Método de resolução sem while
*  utilizando "for" com "range"

```python
dados = input('Informe a notas de 0 a 10 separadas por espaços: ')
notas = dados.split()

for i in range(len(notas)):
  notas[i] = float(notas[i])
media = sum(notas) / len(notas)
print(f'A média das {len(notas)} notas é {media:.1f}')

quadrado = []

for i in range(len(notas)): # percorra todos os range em i. contagem de números como índice
  quadrado.append((notas[i]-media)**2)
# cria lista quadrado nota menos media elevado ao quadrado para cada nota.

desvio = (sum(quadrado) / (len(notas)-1))**0.5
# desvio recebe soma do quadrado divido por quantidade de notas menos 1 elevado a meio.

print(f'O Desvio Padrão das notas {len(notas)} notas é {desvio:.2f}')
```

*  Elabore um programa que que receba um número inteiro do usuário e verifique
se o número informado é par ou ímpar, se for par imprima uma mensagem dizendo
que o número é par e encerre o loop, caso seja ímpar exiba uma mensagem
informando que o número não é par e peça para o usuário digitar outro número,
até que o número digitado pelo usuário seja par.
* Com Looping

```python
while True: # enquanto verdadeiro, continuar pedindo input
  num = int(input('Digite um número inteiro: '))

  if num % 2 == 0:
    print(f'O {num} é par')
    break
  else:
    print(f"o {num} não é par, digite novamente: ")
```

* Sem Looping

```python
n = int(input('digite um número inteiro: '))
if n % 2 == 0: # o resto da divisão é zero
  print(f'o número {n} é um número par')
else:
  print(f'o número {n} é um número ímpar')
```

##Exemplo: Função DEF

*  Criando funções definidas pelo usuário
    * A palavra-chave def é usada para iniciar a definição de uma função, seguida pelo nome da função e parênteses vazios (). Se a função aceitar parâmetros, esses parâmetros seriam listados dentro dos parênteses. O corpo da função é indentado e contém o código que será executado quando a função for chamada.
*  Função para calcular o índice de massa corporal

```python
def imc(peso, altura):
# def - palavra reservada para definição de uma função.
# cria uma função com essa definição (def).
# argumento somente para coisas prontas
  imc = peso/altura**2
  return imc
```

```python
douglas = imc(80, 1.85)
print(f"O IMC de Douglas é {douglas:.2f}")
```
