##6.2 - Removendo/renomeando caracteres das colunas

###6.2.1 - Removendo caracteres do custo
* símbolo $
* " , " (vírgula)
* transformando em float

```python
# Remove currency symbol
df['custo'] = df['custo'].str.replace('$', '')

# Substituir vírgulas por pontos na coluna 'custo'
df['custo'] = df['custo'].str.replace(',', '')

# Converter a coluna para float
df['custo'] = df['custo'].astype(float)
```

###6.2.2 - Removendo caracteres de duração

```python
# Remove days de duracao
df['duracao'] = df['duracao'].str.replace(' days', '')

# Converter a coluna para float
df['duracao'] = df['duracao'].astype(float)
```

###6.2.3 - Renomeando palavras no publico_alvo

```python
# renomeando all ages para todas_as_idades em publico_alvo
df['publico_alvo'] = df['publico_alvo'].str.replace('All Ages', 'todas_as_idades')
# renomeando men para homem em público_alvo
df['publico_alvo'] = df['publico_alvo'].str.replace('Men', 'homem')
# renomenando women para mulher em público_alvo
df['publico_alvo'] = df['publico_alvo'].str.replace('Women', 'mulher')
```

##6.5 - Confirmação de Dados da Verificação

```python
'''
Tipos de validação:
object=String

:pa.Column(pa.String)
:pa.Column(pa.DateTime)
:pa.Column(pa.Int)
:pa.Column(pa.Float)
'''
```

```python
schema = pa.DataFrameSchema({'id':pa.Column(pa.Int),
                             'empresa':pa.Column(pa.String),
                             'campanha':pa.Column(pa.String),
                             'publico_alvo':pa.Column(pa.String),
                             'duracao':pa.Column(pa.Float),
                             'canal':pa.Column(pa.String),
                             'taxa_conversao':pa.Column(pa.Float),
                             'custo':pa.Column(pa.Float),
                             'retorno_investimento':pa.Column(pa.Float),
                             'localizacao':pa.Column(pa.String),
                             'lingua':pa.Column(pa.String),
                             'cliques':pa.Column(pa.Int),
                             'impressao':pa.Column(pa.Int),
                             'engajamento':pa.Column(pa.Int),
                             'segmento':pa.Column(pa.String),
                             'data':pa.Column(pa.DateTime),
                             'ano':pa.Column(pa.String),
                             'mes':pa.Column(pa.String),
                             'dia':pa.Column(pa.String),

                              })
```
